```python
from google.colab import userdata
from qdrant_client import QdrantClient
client = QdrantClient(url=userdata.get("QDRANT_URL"),api_key=userdata.get("QDRANT_API_KEY"))

collections = client.get_collections()

print(f"Connected to Qdrant Cloud {len(collections.collections) }Collections")
```

```python
from qdrant_client import models
collection_name = "First_Collections"
client.create_collection(
    collection_name = collection_name,
    vectors_config = models.VectorParams(
        size = 4,
        distance = models.Distance.COSINE
    )
)
```

```python
#declare the points
points = [
    models.PointStruct(
        id = 1,
        vector = (0.1,0.2,0.3,0.4),
        payload = {"Language":"Spanish"}
    ),
    models.PointStruct(
        id = 2,
        vector = (1.1,1.2,1.3,1.4),
        payload ={"Language":"Hindi"}
    ),
    models.PointStruct(
        id = 3,
        vector = (2.1,2.2,2.3,2.4),
        payload = {"Language":"Russian"}
    )
]
client.upsert(
    collection_name = collection_name,
    points = points
)

```

```python
collection_info = client.get_collection(collection_name)
print("Collections Info: ",collection_info)
```

```python
query_vector = [0.1,0.2,0.3,0.4]
search_result = client.query_points(
    collection_name = collection_name,
    query = query_vector,
    limit = 1
)
print("Results : ",search_result)
```

```python
query_vector2 = [1.1,1.2,1.3,1.4]
search_result = client.query_points(
    collection_name = collection_name,
    query = query_vector2,
    limit = 1
)
print("search_result",search_result)
```

```python
query_vector3 = [3.1,3.2,3.3,3.4]

search_result = client.query_points(
    collection_name = collection_name,
    query = query_vector3,
    limit = 1

)
print("Search Result: ",search_result)
```
